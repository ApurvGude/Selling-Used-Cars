# EDA and Feature Engineering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
pd.set_option('display.max_columns', None) 
import glob
import plotly.express as px
import plotly.graph_objects as go
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder,StandardScaler
import plotly.io as pio
pio.renderers.default = "svg"
```

## Loading the data

```python
files=  glob.glob('data/raw/*.csv')
arr = []
for i in range(len(files)):
    data = pd.read_csv(files[i])
    data['Company'] = files[i].split('\\')[1].split('.')[0]
    arr.append(data)
data = pd.concat([arr[i] for i in range(len(files))])    
```

## EDA (Exploratory Data Analysis) of Data

```python
a = data['Company'].value_counts()
b = data.groupby('Company')['price'].mean()
c = pd.DataFrame([a,b]).T.sort_values(by = 'Company',ascending=True)
fig = px.bar(c,y  =c.index,x =  'Company',color = 'price',orientation='h',text = c.index,color_continuous_scale =px.colors.sequential.Viridis)
fig.update_traces(textposition='inside',insidetextanchor='start',textfont_size=14,textfont_color='black')
fig.update_yaxes(showticklabels=False)
fig.update_layout(
title = 'Most common car companies and their average car price',
xaxis_title = 'Number of cars sold',
yaxis_title ='Car Company',
)
```

```python
# Adding a feature to the dataset depicting the age of the car
a = pd.DataFrame(data.groupby(['model','year'])['price'].mean())
a['Age']=0
a = a.reset_index()

for i in np.unique(a['model']):
    q = np.array(a.loc[a['model']==i,'year'])[0]
    for j in a.loc[a['model']==i].index:
        a.loc[j,'Age']= a.loc[j,'year']-q
b = a.drop(['price'],axis=1)
data = pd.merge(data,b,how='left',left_on = ['model','year'],right_on=['model','year'])        
```

```python
a = data.groupby(['model','Age'])['price'].sum()
a = a.reset_index()
p = a['model'].value_counts()[-50:].index
for i in p:
    q = a.loc[a['model']==i].index
    a = a.drop(q,axis=0)
```

```python
# Depicitng the changing selling prices of 5 random models over the years
r = np.unique(a['model'])
fig,ax = plt.subplots(figsize=(10,8))
for i in range(5):
    h = np.random.randint(0,len(r))
    t = a.loc[a['model']==r[h]]
    sns.lineplot(x= t['Age'],y =(t['price']),ax=ax)    
```

```python
w = data['Age'].value_counts()
fig = px.bar(x = w.index,y = w)
fig.update_layout(
xaxis_title = 'Number of used cars sold',
yaxis_title ='Age of sold car',
title = 'Number of cars sold at different ages')
```

```python
a = pd.DataFrame(data.groupby('Age')['mileage'].mean())
px.bar(a,x=a.index,y = 'mileage',title = 'Average mileage of car as age of car increases')
```

## Feature Engineering

```python
# Removing NaN values and cleaning data
g = data.loc[(data['tax'].isna()) & (data['tax(£)'].isna()==False)].index
data.loc[g,'tax'] = data.loc[g,'tax(£)']*1.3

data = data.drop(['tax(£)'],axis=1)

d = data.loc[(data['model']==' Focus'),'tax'].mean()
data['tax'] = data['tax'].fillna(d)
e = data.loc[(data['model']==' Focus'),'mpg'].mean()
data['mpg'] = data['mpg'].fillna(e)

data['tot_price'] = data['price']+data['tax']
data = data.drop(['price','tax'],axis=1)

x = data.drop(['tot_price','year'],axis=1)
y = data['tot_price']

# Label encoding columns with categorical features and normalizing numerical columns
cols = ['model','transmission','fuelType','Company']
cols1 = ['mileage','mpg']
ct = ColumnTransformer(transformers = [('categorical_features',OrdinalEncoder(),cols),('num_features',StandardScaler(),cols1)],remainder = 'passthrough')

x1 = pd.DataFrame(ct.fit_transform(x))
x1 = x1.rename(columns = {0:'model',1:'transmission',2:'fuelType',3:'Company',4:'mileage',5:'mpg',6:'engineSize',7:'Age'})
```

```python
#Savling the ct model to be used in next notebook
import pickle

with open('ct_model.pkl', 'wb') as file:  
    pickle.dump(ct, file)
```

## Saving the processed data

```python
x1.to_csv('data/processed/processed_x.csv')
pd.DataFrame(y).to_csv('data/processed/processed_y.csv')
data['start_year'] = data['year']-data['Age']
data.to_csv('data/processed/input_data.csv')
```
